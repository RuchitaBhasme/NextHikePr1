```python
q1=concat_data['cnt'].quantile(.25)
q3=concat_data['cnt'].quantile(.75)
```

```python
def outlier(Dropdata1_2,cols):
    q1=Dropdata1_2[cols].quantile(.25)
    q3=Dropdata1_2[cols].quantile(.75)
    iqr=q3-q1
    min_r=q1-1.5*iqr
    max_r=q3+1.5*iqr
    outlier_indices=Dropdata1_2.index[(Dropdata1_2[cols]<min_r)|(Dropdata1_2[cols]>max_r)]
    return outlier_indices
```
